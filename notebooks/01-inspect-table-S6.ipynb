```python
import pandas as pd

# Path to the Excel file
excel_file_path = '../data/rawdata.xlsx'

print("--- Attempting to correctly read sheet: Table S6 ---")

try:
    # We will tell pandas to use the VERY FIRST ROW as the header (header=0)
    df_s6 = pd.read_excel(excel_file_path, sheet_name='Table S6', header=0)

    # Clean up column names and drop empty rows
    df_s6.columns = df_s6.columns.str.strip()
    df_s6.dropna(how='all', inplace=True)

    print("\nSuccessfully loaded 'Table S6'.")
    print("\nColumns:", df_s6.columns.tolist())
    print("\nData Preview:")
    print(df_s6.head())

except Exception as e:
    print(f"An error occurred: {e}")
```
